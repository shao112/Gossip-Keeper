```python
from bs4 import BeautifulSoup 
import requests 
import re

# 掛上headers模擬使用者讀取網頁的行為
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'}
 # 目標網站 
url = "https://www.pttweb.cc/hot/all/today"
 # 發出 get請求 
resp = requests.get(url, headers=headers) 
 # 網頁編碼 
resp.encoding = 'utf-8' 
 # 透過 BeautifulSoup 進行網頁解析 
soup = BeautifulSoup(resp.text, 'html.parser') 

#print(soup.prettify())輸出網頁碼

divs = soup.find_all('div', 'e7-right-top-container e7-no-outline-all-descendants')

href_list = []

# 換個網頁爬文章內容 
root = 'https://www.ptt.cc' 
for div in divs: 
    #文章連接
    link = div.find_all('a','e7-article-default')
    #print(link)
    for link1 in link:
        link2= link1.get('href')
        href_list.append(link2)
#print(href_list)


#0917
for href_single in href_list:
    url2='https://www.pttweb.cc'+ href_single
    print(url2)
    terms ={'q':'網路爬蟲'}
    html2 = requests.get(url2, params = terms)

    #status_code是回應狀態碼，如果狀態碼為200時，表示連線成功。
    if html2.status_code ==200:
        #print('成功哈哈')
        html2.encoding = 'utf-8' 
     # 透過 BeautifulSoup 進行網頁解析 
        soup2 = BeautifulSoup(html2.text, 'html.parser') 

        content = soup2.prettify()
        title = soup2.title.string
        #拿掉前面分類
        if title[0:4]=="Re: ":
            title = re.sub(r'.', '', title, count = 9)
        else:
            title = re.sub(r'.', '', title, count = 5)

        #拿掉特殊符號
        specialChars = "】!#$%^&*()[]／?？!！「」.：'\u3000'" 
        for specialChar in specialChars:
            title = title.replace(specialChar, '')
            #print(title)
        print(title)

        ret = soup2.find_all(attrs={'data-hid':'og:description'})
    #print(ret)
    #print(type(ret))
        con = []
        for res in ret:
            con1= res.get('maintextimageurlcontent')
            con.append(con1)
        print(con)
    else:
        print('失敗q')
print('Done')
```
